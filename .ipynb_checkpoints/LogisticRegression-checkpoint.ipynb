### __Text cleaning__ from stopwords,links and pancutation.

Creating a set of stopwords first consisted of nltk library and wordcloud library:

````python
from wordcloud import STOPWORDS as stopwords1,WordCloud
from nltk.corpus import stopwords as stopwords2
from nltk import word_tokenize
import string

stopwords = set(stopwords1)
stopwords_nltk =  set(stopwords2.words('english'))
stopwords.update(stopwords_nltk)

# Removes stopwords from sentences
def removeStopWords(text):
    return ' '.join([y for y in word_tokenize(text) if y not in stopwords])
````

### __Lemmatization and usefull word processing functions__


````python
import unicodedata
import re
import string

# Function that lemmatizes wanted columns
lemmatizer = nltk.stem.WordNetLemmatizer()
def Lemmatize(text):
    return ' '.join([lemmatizer.lemmatize(y,'v') for y in word_tokenize(text)])

def normalazation(text):
    return ' '.join([normalize_word(y) for y in word_tokenize(text)])

def normalize_word(word):
    # Will remove the encode token by token
    word = unicodedata.normalize('NFKD', word).encode('ASCII','ignore').decode('ASCII')
    return word

def smallwordsRemoval(text):
    return ' '.join([remove_smallWords(y) for y in word_tokenize(text)])

def remove_smallWords(word):
    if (len(word) > 2):
        return word
    else:
        return ''
````

````python
tweetDf['tweet'] = tweetDf['tweet'].str.lower().apply(Lemmatize).apply(removeStopWords)
tweetDf.head(5)
````

Creating columns with or without __pancutation__.
Clearing all __links__ and tags.

````python
tweetDf['clean comments (no links)'] = tweetDf['tweet'].str.replace(r'@\S+',' ',regex=True).replace(r'\\\S+',' ',regex=True).replace('\n',' ')
tweetDf['clean comments (no links & pancutation)'] = tweetDf['clean comments (no links)']
tweetDf['clean comments (no links & pancutation)'] = tweetDf['clean comments (no links & pancutation)'].str.replace('[{}]'.format(string.punctuation), '')
tweetDf.head(5)
````

__Vectorazation__

````python
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
from sklearn.model_selection import train_test_split

# Vectorizes train and test
def vectorizeData(train,test,vectorizer):
    train = vectorizer.fit_transform(train)
    test = vectorizer.transform(test)
    return train,test
````

### POS tagging
> The Natural Language Toolkit (NLTK) is a platform used for building programs for text analysis. One of the more powerful aspects of the NLTK module is the Part of Speech tagging.





````python
# Returns a list of [(word,tag),...] for a given string
def pos_preprocess(text):
    return nltk.pos_tag(nltk.word_tokenize(text))

# This function concatenates a comment with the tags prodused and makes a new column only from the tags 
def addPosTags(listOfText):
    pos_tags_column = []
    pos_tags_only = []
    for text in listOfText:
        pos_tags_list = []
        doc = pos_preprocess(text)
        for token in doc:
            pos_tags_list.append(token[1])
        all_pos_tags = ' '.join(pos_tags_list)
        pos_tags_only.append(all_pos_tags)
        newtxt = text + ' ' + all_pos_tags
        pos_tags_column.append(newtxt)
    return pos_tags_column,pos_tags_only
````

````python
tweetDf['POS_comment'],tweetDf['POS_tags'] = addPosTags(tweetDf['clean comments (no links & pancutation)'])
tweetDf.head(5)
````

### Best clean comments
The function above cleans even more and better the tweets and it will be used for better classification.

````python
def cleanText(df,column_toSave,initial_column):
    df[column_toSave] = df[initial_column].apply(normalazation).apply(smallwordsRemoval).apply(Lemmatize).apply(removeStopWords)

cleanText(tweetDf,'Best-clean-tweets','clean comments (no links & pancutation)')
tweetDf.head(5)
````

### Valence Aware Dictionary and sEntiment Reasoner - __VADER__
> VADER (Valence Aware Dictionary and sEntiment Reasoner) is a lexicon and rule-based sentiment analysis tool that is specifically attuned to sentiments expressed in social media. VADER uses a combination of A sentiment lexicon is a list of lexical features (e.g., words) which are generally labelled according to their semantic orientation as either positive or negative. 

````python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')

def VaderperWord(text):
    newlist = []
    vader = SentimentIntensityAnalyzer()
    for word in word_tokenize(text):
        vresult = vader.polarity_scores(word)
        resultmax = max(vresult,key= lambda x : vresult[x])
        newlist.append(resultmax)
         
    return ' '.join([y for y in newlist])
````

````python
tweetDf['Vader sentences'] = tweetDf['clean comments (no links & pancutation)'].apply(VaderperWord)
````

## WordClouds

--- 


For a better evaluation of data

````python
import collections
# Function that given a list of data will produce a WordCloud
def generate_WordCloud(data,stopwords):
    textWords = []
    for x in data:
        for y in word_tokenize(x) :
            y = y.lower()
            y = WordNetLemmatizer().lemmatize(y)
            textWords.append(y)
                
    counter = collections.Counter(textWords)
    words = WordCloud(background_color ='white',stopwords=stopwords,margin=0).generate_from_frequencies(frequencies=dict(counter))

    plt.figure(figsize=(10,5))
    plt.imshow(words,interpolation='bilinear')
    plt.axis("off")
    plt.show()
````

### Confusion matrix


````python
import itertools 

# Function that creates a confusion matrix
def create_ConfusionMatrix(confusionMatrix,title):
    classes = ['Negative','Positive']
    cmap = plt.cm.Reds

    plt.imshow(confusionMatrix, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    thresh = confusionMatrix.max() / 2.
    for i, j in itertools.product(range(confusionMatrix.shape[0]), range(confusionMatrix.shape[1])):
        plt.text(j, i, confusionMatrix[i, j],horizontalalignment="center",color="white" if confusionMatrix[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
    plt.ylim([1.5, -.5])
    plt.show()  
````

### Classification function

Function ```makeClassification``` makes the classification,in these three steps:
- GridSearch for hyperparameter tunning
- With the best hyperparameters makes classification
- Prints scores
- Makes Confusion matrix



````python
import warnings
import sklearn.exceptions
warnings.filterwarnings("ignore", category=sklearn.exceptions.FitFailedWarning)
warnings.filterwarnings("ignore", category=UserWarning)
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.model_selection import StratifiedKFold
folds = StratifiedKFold(n_splits=5)
param_grid = {'penalty': ['l1', 'l2', 'elasticnet'],
              'C': [0.1, 1 ,10],
              'solver': ['newton-cg','lbfgs','liblinear','sag']}  

def makeClassification(X_train, X_test, y_train, y_test):
    
    LR_model = LogisticRegression()
    grid = GridSearchCV(LR_model,param_grid,refit=True,cv=folds)
    grid.fit(X_train, y_train)
    grid_predictions = grid.predict(X_test)
    print("GridSearch ended")
    print("Best parameters: ",grid.best_params_)
    print("Running LogisticRegression again with these parameters")
    LR_model = LogisticRegression(**grid.best_params_)
    LR_model.fit(X_train, y_train)
    prediction = LR_model.predict(X_test)
    print("Scores:")
    print("-----------------------------------------------------------")
    print(classification_report(y_test, prediction, target_names=['Negative','Positive']))
    print("-----------------------------------------------------------")
    cm = sklearn.metrics.confusion_matrix(y_test,prediction,labels=[0,1])
    create_ConfusionMatrix(cm,'Confusion matrix')
````
